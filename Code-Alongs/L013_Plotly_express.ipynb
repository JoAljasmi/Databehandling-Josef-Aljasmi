# Plotly Express

```python
import plotly_express as px
```

```python
df = px.data.gapminder()
df.head()
```

```python
px.strip(df.query("year==2007"), x="lifeExp", y="continent", color="continent", hover_name="country", hover_data={"continent" : False, "pop":True})
```

```python
px.strip(df, x="lifeExp", color="continent", hover_name="country", facet_col="year", facet_col_wrap=4)
```

```python
px.strip(df, x="lifeExp", color="continent", hover_name="country", animation_frame="year", range_x=[10, 90])
```

```python
px.histogram(df.query("year==2007"), x="lifeExp", y= "pop", color="continent", marginal="rug") #Marginal can have "box", "rug" or "violin"
```

```python
px.bar(df.query("year==2007"), x="pop", y="continent", color="lifeExp", hover_name="country")
```

```python
px.sunburst(df.query("year==2007"), values="pop", color="lifeExp", path=["continent", "country"], hover_name="country")
```

```python
px.choropleth(df.query("year==2007"), color= "lifeExp", locations="iso_alpha", hover_name="country")
```

```python
px.scatter(df.query("year==2007"), 
    x= "gdpPercap", y= "lifeExp",
    range_x=[100,100000], range_y=[25,90],
    animation_frame="year", animation_group="country",
    color= "continent",
    size= "pop", size_max=70, 
    log_x=True,
    hover_name="country")
```
